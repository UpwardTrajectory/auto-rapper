```python
import gpt_2_simple as gpt2

model_name = "117M"
# gpt2.download_gpt2(model_name=model_name)   # model is saved into current directory under /models/117M/
```

```python
import json
with open('artists/eminem.json') as f:
    raps = json.load(f)
```

```python
rap_titles = list(raps.keys())
```

```python
corpus = ""

for song, lyrics in raps.items():
    corpus +=  ' \n\n\n ' + song + ' \n\n '
    corpus += lyrics.replace('\n', ' \n')

    
```

```python
sess = gpt2.start_tf_sess()
gpt2.finetune(sess,
              "corpus.txt",
              model_name=model_name,
              learning_rate=0.01,
              steps=5)   # steps is max number of training steps

gpt2.generate(sess)
```
