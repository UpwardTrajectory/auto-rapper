```python
token = 'for_one_time_use_paste_your_token_here'

path = "/Users/dakaspar/flatiron/projects/auto-rapper/.secret/api_key.json"
rappers = {}

with open(path) as f:
    temp_api = json.load(f)
    secret = temp_api['secret']
    api_id = temp_api['id']
    token = temp_api['token']

genius = lyricsgenius.Genius(token)
```

```python
def get_lyrics(artist, max_songs=None, json_folder='default'):
    """
    Search genius.com for {max_songs} songs by {artist} and return a dictionary of the {song_title : lyrics_as_single_string}.
    optionally, save the dictionary as `artist.json` inside the path: json_folder='destination_folder_path'
    
    by default, will search for ALL available songs (sorted descending by popularity),  and save them in `./artists/` folder
    """
    current_artist = genius.search_artist(artist, max_songs=max_songs, get_full_info=False)
    artist_name = current_artist.name.lower().translate(str.maketrans('', '', string.punctuation)).replace(" ", "_")

    all_lyrics = {}
    for song in current_artist.songs:
        all_lyrics[song.title] = song.lyrics

    if json_folder is not None:
        if json_folder is 'default':
            json_folder = '/Users/dakaspar/flatiron/projects/auto-rapper/artists/' #Folder you want to save your lyrics to
    
        json_folder += artist_name + '.json'
        
        with open(json_folder, 'w') as fp:
            json.dump(all_lyrics, fp)
            
    if all_lyrics not in rappers.values():
        rappers[artist_name] = all_lyrics
    return all_lyrics
```

```python
to_do = ['Snoop Dogg']

for artist in to_do:
    get_lyrics(artist)
```
